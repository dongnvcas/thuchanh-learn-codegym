```python
#[Bài tập] Nâng cấp ngôi nhà mơ ước
import turtle
h = turtle.Turtle()
h.speed(10) # 1:slowest, 3:slow, 5:normal, 10:fast, 0:fastest
h.pencolor("lightblue")
h.penup()
h.goto(-100,0)
h.pendown()
#ve than nha
h.fillcolor("lightblue")
h.begin_fill()
for i in range(2):
  h.fd(200)
  h.rt(90)
  h.fd(100)
  h.rt(90)
h.end_fill()
#Ve mai nha
h.fillcolor("red")
h.begin_fill()
h.lt(45)
h.fd(100)
h.rt(90)
h.fd(100)
h.end_fill()
#ve ong khoi
h.fillcolor("brown")
h.begin_fill()
h.bk(60)
h.lt(135)
h.fd(30)
h.rt(90)
h.fd(20)
h.rt(90)
h.fd(50)
h.end_fill()
#ve khoi
h.pencolor("grey")
h.penup()
h.rt(180)
h.fd(70)
h.pendown()
h.fd(50)
h.penup()
h.lt(90)
h.fd(20)
h.rt(90)
h.bk(40)
h.pendown()
h.fd(30)
#ve cua
h.fillcolor("yellow")
h.penup()
h.goto(-20,-100)
h.pendown()
h.begin_fill()
for i in range(2):
  h.fd(80)
  h.rt(90)
  h.fd(40)
  h.rt(90)
h.end_fill()
# Ve cay so 1
t = turtle.Turtle()
t.color("green")
t.speed(5)
t.penup()
t.goto(-200,-150)
t.pendown()
#Ve than cay
t.fillcolor("lightgreen")
t.begin_fill()
t.lt(80)
t.fd(80)
t.rt(80)
t.fd(20)
t.rt(80)
t.fd(80)
t.rt(100)
t.fd(50)
t.end_fill()
#Ve tan cay 1
t.penup()
t.goto(-225,-71)
t.rt(180)
t.pendown()
t.fillcolor("green")
t.begin_fill()
t.fd(100)
t.lt(135)
t.fd(70.71)
t.lt(90)
t.fd(70.71)
t.lt(135)
t.end_fill()
# Ve tan cay 2
t.penup()
t.goto(-212,-21)
t.pendown()
t.fillcolor("blue")
t.begin_fill()
t.fd(80)
t.lt(135)
t.fd(56.5685424949)
t.lt(90)
t.fd(56.5685424949)
t.lt(135)
t.end_fill()
# Ve tan cay 3
t.penup()
t.goto(-202,19)
t.pendown()
t.fillcolor("orange")
t.begin_fill()
t.fd(60)
t.lt(135)
t.fd(42.4264068712)
t.lt(90)
t.fd(42.4264068712)
t.lt(135)
t.end_fill()
```
